```python
# Setting up dictionaries
feats = defaultdict(lambda: np.array)
labels = defaultdict(lambda: defaultdict(lambda: np.array))
dists = defaultdict(lambda: np.array)
edges = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: defaultdict(set))))
edge_nums = defaultdict(lambda: 1)

target_types = defaultdict(set)
relations = defaultdict(lambda: defaultdict(str))
feat_dims = defaultdict(int)
label_dims = defaultdict(lambda: defaultdict(int))
```

```python
for file in os.listdir(data_dir):
    if file.endswith(".npy") and 'nodes-' in file:
        print(f"file: {file}")

        with open(os.path.join(data_dir, file), 'rb') as f:
            print(f"f: {f}")
            
            _type = re.search('nodes-(.*?).npy', file).group(1)
            print(f"_type: {_type}")
            
            feats[_type] = numpy.load(f).astype(numpy.float64)
            #print(f"feats[{_type}]: {feats[_type]}")
            
            feat_dims[_type] = feats[_type].shape[1]
            #print(f"feat_dims[{_type}]: {feat_dims[_type]}")
            
            if _type in ('paper', 'author', 'venue'):
                
                labels[_type]['L1'] = numpy.load(f).astype(numpy.int64)
                print(f"labels[{_type}]['L1']: {labels[_type]['L1']}")
                print(f"labels[{_type}]['L1']: {labels[_type]['L1'].shape}")
                
                label_dims[_type]['L1'] = numpy.load(f).astype(numpy.int64).item()
                print(f"label_dims[{_type}]['L1']: {label_dims[_type]['L1']}")
                
                dists[_type] = numpy.load(f).astype(numpy.int64)
                print(f"dists[{_type}]: {dists[_type]}")
                print(f"dists[{_type}]: {dists[_type].shape}")
                
                labels[_type]['L2'] = numpy.load(f).astype(numpy.int64)
                print(f"labels[{_type}]['L2']: {labels[_type]['L2']}")
                print(f"labels[{_type}]['L2']: {labels[_type]['L2'].shape}")
                
                label_dims[_type]['L2'] = numpy.load(f).astype(numpy.int64).item()
                print(f"label_dims[{_type}]['L2']: {label_dims[_type]['L2']}")
                
            if _type == 'paper':
                labels[_type]['venue'] = numpy.load(f).astype(numpy.int64)
                label_dims[_type]['venue'] = numpy.load(f).astype(numpy.int64).item()
            
            print("")
```
